```python
#HISTOGRAMA 1 - Costo Total del Proyecto Escenario 1
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Ruta del archivo fuente y de salida
ruta_archivo = r"C:\Users\Asus2\Desktop\20250521_Datos_Simulados.xlsx"
ruta_salida = r"C:\Users\Asus2\Downloads\Histograma_Escenario_1_Costos.png"
```

```python
# Leer datos
df = pd.read_excel(ruta_archivo, sheet_name='Datos_Simulados')
```

```python
# Seleccionar columna del escenario 1
columna_costos_e1 = 'Modelo_Costo!$B$20'
df[columna_costos_e1] = pd.to_numeric(df[columna_costos_e1], errors='coerce')
data = df[columna_costos_e1].dropna()
```

```python
# Calcular percentiles
p10, p50, p90 = np.percentile(data, [10, 50, 90])

# Crear figura
fig, ax1 = plt.subplots(figsize=(14, 6))

# Histograma
counts, bins, _ = ax1.hist(data, bins=30, density=True, alpha=0.7, color='lightseagreen', edgecolor='black')
ax1.set_xlabel('Costo total del proyecto')
ax1.set_ylabel('% de simulaciones por costo')
ax1.set_title('Costo Total del Proyecto – Escenario 1 (Secuencial A → B → C)')

# Percentiles
for perc, value in zip(['P10', 'P50', 'P90'], [p10, p50, p90]):
    ax1.axvline(value, color='blue', linestyle='dashed')
    ax1.text(value, max(counts)*0.9, f'{perc}\n{value:,.0f}', color='black', ha='center')

# CDF
ax2 = ax1.twinx()
sorted_data = np.sort(data)
cdf = np.arange(len(sorted_data)) / float(len(sorted_data))
ax2.plot(sorted_data, cdf, color='navy')
ax2.set_ylabel('Probabilidad acumulada')
```

```python
# Exportar y mostrar
fig.tight_layout()
fig.savefig(ruta_salida, dpi=300)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# Rutas
ruta_archivo = r"C:\Users\Asus2\Desktop\20250521_Datos_Simulados.xlsx"
ruta_salida = r"C:\Users\Asus2\Downloads\Histograma_Escenario_1_FechaFinal.png"
```

```python
# Leer datos
df = pd.read_excel(ruta_archivo, sheet_name='Datos_Simulados')
```

```python
# Seleccionar columna de duración del escenario 1
columna_fecha_e1 = 'Modelo_Duracion!$E$11'
df[columna_fecha_e1] = pd.to_numeric(df[columna_fecha_e1], errors='coerce')
data = pd.to_datetime(df[columna_fecha_e1], unit='D', origin='1899-12-30').dropna()
```

```python
# Calcular percentiles (conversión a enteros de nanosegundos para cálculos)
data_int = data.astype('int64')
p10, p50, p90 = np.percentile(data_int, [10, 50, 90])
p10_dt, p50_dt, p90_dt = pd.to_datetime([p10, p50, p90])

# Crear figura
fig, ax1 = plt.subplots(figsize=(14, 6))

# Histograma
counts, bins, _ = ax1.hist(data, bins=30, density=True, alpha=0.7,
                           color='moccasin', edgecolor='orange')
ax1.set_xlabel('Fecha de finalización del proyecto')
ax1.set_ylabel('% de simulaciones por fecha')
ax1.set_title('Fecha de Finalización del Proyecto – Escenario 1 (Secuencial)')

# Eje X como fechas
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
ax1.xaxis.set_major_locator(mdates.AutoDateLocator())

# Líneas de percentiles
for perc, dt in zip(['P10', 'P50', 'P90'], [p10_dt, p50_dt, p90_dt]):
    ax1.axvline(dt, color='blue', linestyle='dashed')
    ax1.text(dt, max(counts) * 0.9, f'{perc}\n{dt.strftime("%d-%m-%Y")}', color='black', ha='center')

# Curva acumulada (CDF)
ax2 = ax1.twinx()
sorted_data = np.sort(data)
cdf = np.arange(len(sorted_data)) / float(len(sorted_data))
ax2.plot(sorted_data, cdf, color='navy')
ax2.set_ylabel('Probabilidad acumulada')
```

```python
#HISTOGRAMA 7 - DIAGRAMA DE DISPERSIÓN
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Rutas
ruta_archivo = r"C:\Users\Asus2\Desktop\20250521_Datos_Simulados.xlsx"
ruta_salida = r"C:\Users\Asus2\Downloads\Scatter_Plot.png"
```

```python
# Cargar y preparar datos
df = pd.read_excel(ruta_archivo, sheet_name="Datos_Simulados")
df = df.rename(columns={
    'Modelo_Duracion!$F$10': 'Duracion_E2_Total',
    'Modelo_Costo!$C$20': 'Costo_E2_Total'
})
df['Duracion_E2_Total'] = pd.to_numeric(df['Duracion_E2_Total'], errors='coerce')
df['Costo_E2_Total'] = pd.to_numeric(df['Costo_E2_Total'], errors='coerce')
df = df.dropna(subset=['Duracion_E2_Total', 'Costo_E2_Total'])
```

```python
# Calcular porcentajes por cuadrante
umbral_duracion = 100
umbral_costo = 20000
total = len(df)

cuadrantes = {
    "Bajo Costo, Baja Duración": ((df['Duracion_E2_Total'] < umbral_duracion) & (df['Costo_E2_Total'] < umbral_costo)).sum(),
    "Alto Costo, Baja Duración": ((df['Duracion_E2_Total'] < umbral_duracion) & (df['Costo_E2_Total'] >= umbral_costo)).sum(),
    "Bajo Costo, Alta Duración": ((df['Duracion_E2_Total'] >= umbral_duracion) & (df['Costo_E2_Total'] < umbral_costo)).sum(),
    "Alto Costo, Alta Duración": ((df['Duracion_E2_Total'] >= umbral_duracion) & (df['Costo_E2_Total'] >= umbral_costo)).sum()
}

cuadrantes_pct = {k: round(v / total * 100, 1) for k, v in cuadrantes.items()}

# Gráfico de dispersión con anotaciones
fig, ax = plt.subplots(figsize=(10, 6))

ax.scatter(df['Duracion_E2_Total'], df['Costo_E2_Total'], alpha=0.5, color='grey', edgecolor='none')
ax.axvline(umbral_duracion, color='dimgray', linestyle='--', label='Duración = 100 días')
ax.axhline(umbral_costo, color='dimgray', linestyle='--', label='Costo = $20.000')

ax.text(umbral_duracion - 5, umbral_costo - 2000, f"{cuadrantes_pct['Bajo Costo, Baja Duración']}%", fontsize=10, ha='right')
ax.text(umbral_duracion - 5, umbral_costo + 1000, f"{cuadrantes_pct['Alto Costo, Baja Duración']}%", fontsize=10, ha='right')
ax.text(umbral_duracion + 2, umbral_costo - 2000, f"{cuadrantes_pct['Bajo Costo, Alta Duración']}%", fontsize=10, ha='left')
ax.text(umbral_duracion + 2, umbral_costo + 1000, f"{cuadrantes_pct['Alto Costo, Alta Duración']}%", fontsize=10, ha='left')

ax.set_title("Dispersión – Escenario 2: Duración vs. Costo Total con Cuadrantes")
ax.set_xlabel("Duración total del proyecto (días)")
ax.set_ylabel("Costo total del proyecto")
ax.legend()
ax.grid(False)
```

```python
# Exportar y mostrar
fig.tight_layout()
fig.savefig(ruta_salida, dpi=300)
plt.show()
```
